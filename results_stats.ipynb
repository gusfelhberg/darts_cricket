```python
import pickle
import matplotlib.pyplot as plt

loss = pickle.load(open('rewards_per_episodes.pkl','rb'))
plt.plot([i for i in range(len(loss))], loss)
plt.xlabel('episodes')
plt.ylabel('reward')
plt.show()
```

```python
loss = pickle.load(open('rewards_per_episodes.pkl','rb'))
loss = pd.DataFrame(loss).reset_index()
loss.columns = ['Episode','Reward']
```

```python
import statsmodels.api as sm

fig, ax = plt.subplots()
loss.plot(x='Episode', y='Reward', ax=ax)

model = sm.formula.ols(formula='Reward ~ Episode', data=loss)
res = model.fit()
loss.assign(trend=res.fittedvalues).plot(x='Episode', y='trend', ax=ax)
```
