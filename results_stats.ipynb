```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
```

```python
loss = pickle.load(open('rewards_per_episodes.pkl','rb'))
loss = pd.DataFrame(loss,columns=['episode','agent_reward','agent_score','p1_reward','p1_score'])


fig, ax = plt.subplots()
loss.plot(x='episode', y='agent_reward', ax=ax)

model = sm.formula.ols(formula='agent_reward ~ episode', data=loss)
res = model.fit()
loss.assign(trend=res.fittedvalues).plot(x='episode', y='trend', ax=ax)
```

```python
import pickle
agent_aims_history = pickle.load(open('agent_aims_history.pkl','rb'))
```

```python
def plot_aim_dist(episode):
    pd.DataFrame(agent_aims_history,columns=['Episodes','Number','Multi']).query('Episodes == @episode').iloc[:,1].plot.hist(bins=25,title='Episode '+str(episode),xlim=(0,25))
```
